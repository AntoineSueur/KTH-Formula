```python
import time
import numpy as np
from math import pi, exp, sin
import pandas as pd

from ipywidgets import widgets
from bokeh.models.sources import ColumnDataSource
from bokeh.plotting import figure
from bokeh.io import output_notebook, show, push_notebook
from bokeh.layouts import row, widgetbox, column
from bokeh.models.widgets import Button
import keyboard

output_notebook()

def lambda_func(t):
    return 5*sin(2*pi*1*t)

def h_func(t):
    return 3*pi*exp(-lambda_func(t))

class plot_graph():
       
    def __init__(self):        
        
        self.my_figure = figure(plot_width=900, plot_height=300, title='Experiment')
        self.my_figure.title.align='center'
        self.test_data = ColumnDataSource(data=dict(x=[0], y=[0]))
        self.line = self.my_figure.line("x", "y", source=self.test_data)        
        
        self.layout = row(self.my_figure)
        
        self.handle = show(self.layout, notebook_handle=True)

        self.new_data=dict(x=[0], y=[0])
        self.x = []
        self.y = []
        
        self.step_size = 0.1  # increment for increasing step
        self.max_step = 100  # arbitrary stop point for example
        self.period = 0.1  # in seconds (simulate waiting for new data)
        self.n_show = 1000  # number of points to keep and show
                        
        def start(sender):
            print('start')
            
        def reset(sender):
            print('reset')
            
        def save(sender):
            df = pd.DataFrame({'time':self.x,'h_value':self.y})
            df.to_csv('data.csv')
            
        self.start_button = widgets.Button(description='Start/Resume')
        self.reset_button = widgets.Button(description='Reset')
        self.save_button = widgets.Button(description='Save CSV file')
        
        self.start_button.on_click(start)
        self.reset_button.on_click(reset)  
        self.save_button.on_click(save)  
        
        display(self.start_button,self.reset_button, self.save_button)


        
    def plot(self):
        self.new_data['x'] = self.x  
        self.new_data['y'] = self.y  

        self.test_data.stream(self.new_data, self.n_show)
        push_notebook(handle=self.handle)
        
    def generate_data(self, t):
        self.x.append(t)
        self.y.append(h_func(t))
        
    def reset_data(self):
        self.x = []
        self.y = []
    

    def collect_data(self, func):
        
        print("Press Space to start, Shift to pause and Control to reset")

        t = 0
              
        while(True):        
            if keyboard.is_pressed('control'):  # reset data
                self.reset_data()
                t = 0
            paused = False
            
            if keyboard.is_pressed('space') and not paused:   # Start or resume data collection 
                print('Collecting data...')
                
                while t < self.max_step:

                    if keyboard.is_pressed('control'):   # Pause data collection
                        print("Reset data, press space to start")
                        break
                        
                    if keyboard.is_pressed('shift'):   # Pause data collection
                        print('Paused data collection... press space to continue')
                        paused = True
                        break
                        
                    self.x.append(t)
                    self.y.append(func(t))
                    self.new_data['x'] = self.x = self.x[-self.n_show:]  # prevent filling ram
                    self.new_data['y'] = self.y = self.y[-self.n_show:]  # prevent filling ram
                    self.test_data.stream(self.new_data, self.n_show)

                    push_notebook(handle=self.handle)
                    t += self.step_size
                    time.sleep(self.period)
                    
            elif(keyboard.is_pressed('space') and paused):
                paused = False
                    
```

```python
graph = plot_graph()
graph.collect_data(h_func)
```
